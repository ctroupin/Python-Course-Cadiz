The objective is to show how to organise the plots in subfigures.<br/>
For this illustration, all the subfigures use the same data.

```python
x = np.arange(-5, 5, 0.2)
y = x**2
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Basic subplots

```python
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.subplot(2, 1, 2)
plt.plot(x, -y)
plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.subplot(1, 3, 2)
plt.plot(x, -y)
plt.subplot(1, 3, 3)
plt.plot(x, .5*y)
plt.show()
```

# Customized subplots

Sometimes it's necessary to have more control on the subplot locations.

## Subplot2grid

```python
ax1 = plt.subplot2grid((4,4), (0,0), colspan=3)
ax2 = plt.subplot2grid((4,4), (0,3), rowspan=3)
ax3 = plt.subplot2grid((4,4), (1,0), rowspan=2, colspan=2)
ax4 = plt.subplot2grid((4,4), (1,2), rowspan=2)
ax5 = plt.subplot2grid((4,4), (3,0), colspan=4)
plt.show()
```

For even more controles, see examples from [Subplot2grid](http://matplotlib.org/users/gridspec.html#basic-example-of-using-subplot2grid)

# Axes inset

We can create a figure inside the main figure.

```python
plt.figure(figsize=(7,7))
plt.plot(x, y)
a = plt.axes([.4, .5, .3, .3])
plt.plot(x, -y)
plt.show()
```
