In this notebook we give a few example of how to read and work on an image.<br/>
The main module is [ndimage](http://docs.scipy.org/doc/scipy/reference/ndimage.html) (within scipy), with functions for multi-dimensional image processing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
%matplotlib inline
```

We use an image located in the **figures** directory.<br/>
The variable can be changed to use another image.

```python
imagefile = '../figures/pythoncadiz.jpeg'
```

The [try... except](https://docs.python.org/2/tutorial/errors.html) is useful to handle exceptions:

```python
try:
    imagecadiz = ndimage.imread(imagefile)
except IOError:
    print("File doesn't exist\nCheck image path and name!")
```

Let's examine the type and dimension of the file.

```python
print type(imagecadiz)
print imagecadiz.shape
print imagecadiz.dtype
```

To display an image, there is the [imshow](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow) function. See what happens if we use pcolormesh.

```python
plt.imshow(imagecadiz)
plt.show()
```

# Conversion to gray scale

There are several ways to convert the image to a gray scale image, for example combining the R, G and B channels.<br/> Another possibility is to rely on the Python Imaging Library ([PIL](http://www.pythonware.com/products/pil/)).

```python
from PIL import Image 
```

```python
imagegray = Image.open(imagefile).convert('LA')
plt.imshow(imagegray)
print imagegray
```

# Rotation, zoom, filters...

All these are available in the ndimage package.

## Rotation

```python
imagecadiz_rotate = ndimage.rotate(imagecadiz, 30)
plt.imshow(imagecadiz_rotate)
plt.show()
```

## Blurring

```python
imagecadiz_blurred = ndimage.gaussian_filter(imagecadiz, sigma=5)
plt.imshow(imagecadiz_blurred)
plt.show()
```

## Zoom

```python
imagecadiz_zoom = ndimage.interpolation.zoom(imagecadiz[200:300,100:300,1], 3)
print imagecadiz_zoom.shape
plt.imshow(imagecadiz_zoom)
```
