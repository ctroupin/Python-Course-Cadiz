```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
map = Basemap(llcrnrlon=-10.,llcrnrlat=35.,urcrnrlon=4.,urcrnrlat=43.,
             resolution='l', projection='merc', lat_ts=0.5*(35 + 43))
map.fillcontinents(color='0.85')
map.drawcoastlines()

map.readshapefile('/data_local/DataOceano/ShapeFiles/ESP_adm_shp/ESP_adm2', 'ESP_adm0')

plt.show()
```
