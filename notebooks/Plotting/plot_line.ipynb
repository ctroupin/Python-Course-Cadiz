In this notebook we will a few examples of how to customize a simple plot.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

We create a random vector that we will use for the plots, using the function *rand*.<br/>
The function documentation is also available within the notebook if you type *help(np.random.rand)*.

```python
x = np.random.rand(20)
```

# Simple plot

```python
plt.plot(x)
plt.show()
```

## Changing line aspect

The line width, color, marker and be easily changed by passing optional arguments to the *plot* function.<br/>
We use [*subplot*](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot) to create a 2 X 3 grid of plots in our figure.

```python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot(2,3,1)
plt.plot(x, linewidth=2, color='r')
ax = plt.subplot(2,3,2)
plt.plot(x, linewidth=2, color='r', marker='*')
ax = plt.subplot(2,3,3)
plt.plot(x, linewidth=2, color='r', marker='*', markersize=10)
ax = plt.subplot(2,3,4)
plt.plot(x, 'd-.', linewidth=1, color='b', markersize=7)
ax = plt.subplot(2,3,5)
plt.plot(x, 's--', linewidth=2, color='b', markersize=7, markerfacecolor='g',)
ax = plt.subplot(2,3,6)
plt.plot(x, 'o-', linewidth=2, color='b', markersize=7, markerfacecolor='g', markeredgecolor='y')
plt.show()
```

There are many things you can change in these lines, see [this page](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) for more details.


# Playing with text

## Labels and title

```python
plt.plot(x)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('A random plot')
plt.show()
```

## Legend

Now if we have 2 (or more) variables on the same plot, we want to distinguish them with a [legend](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.legend).

```python
N = 30
x1 = np.arange(0, N)
y1, y2 = np.random.rand(N), np.random.rand(N)
plt.plot(x1, y1, linewidth=2, color='k', label='Variable 1')
plt.plot(x1, y2, '--', linewidth=2, color='r', label='Variable 2')
plt.legend()
plt.show()
```

Sometimes we are not happy with the position of the legend.<br/>
In this case want to have the legend outside of the plot.<br/>
The [Legend guide](http://matplotlib.org/users/legend_guide.html) will help us.

```python
N = 30
x1 = np.arange(0, N)
y1, y2 = np.random.rand(N), np.random.rand(N)
plt.plot(x1, y1, linewidth=2, color='k', label='Variable 1')
plt.plot(x1, y2, '--', linewidth=2, color='r', label='Variable 2')
plt.legend(bbox_to_anchor=(1.05, .25), loc=2, borderaxespad=0.)
plt.show()
```

# A nicer plot

Every element of the plot can be modified, from the font size and style to the colors, grid, aspect, ... 

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.subplot(111)
plt.plot(x1, y1, 'ko--')
plt.xlabel('X', fontsize=20)
plt.ylabel('Variable 1', fontsize=20, rotation=0, horizontalalignment='right')
plt.xticks(color='g', size=20)
plt.yticks(color='b', size=20)
plt.title('Another\n simple plot', fontsize=24, fontweight='bold')
plt.grid()
ax.set_axis_bgcolor('.95')
plt.show()
```

# Using and customizing styles

In order to avoid typing always the same commands and options, there are several possibilities:
* Use of the existing style sheets.
* Define your [own style.](http://matplotlib.org/users/style_sheets.html#defining-your-own-style)
* Modify the default parameters.
   

## Using style sheet

Here is the list of the available styles:

```python
print plt.style.available
```

Let's try the second one.

```python
plt.style.use('bmh')
plt.plot(x)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('A random plot')
plt.show()
```

## Defining your own style

According to the documentation, we have to create a *style file*.<br/>
We create the file **oceanography.mplstyle**.

```python
plt.style.use('./oceanography.mplstyle')
plt.plot(x)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('A random plot')
plt.show()
```

## Modify the default parameters.

There is a file called [**matplotlibrc**](http://matplotlib.org/users/customizing.html#the-matplotlibrc-file) located in the configuration directory (depending on installation) that can be edited (same syntax as the style file).
